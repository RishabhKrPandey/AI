```python
#  for missing values we will do imputation
#  for skewed numerical data - mean or median
#  for categorical data - mode
#  if too many missing values -> either drop it or do some feature engineering
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df['Cabin'].unique()
```

```python
#  as there are too many missing values we will convert cabin -> has_cabin = 0/1 
titanic_df['has_cabin'] = titanic_df['Cabin'].notna().astype(int)
titanic_df.drop('Cabin', axis = 1, inplace=True)
```

```python
titanic_df['has_cabin']
```

```python
#  now we will fill the age with the median value
median = titanic_df['Age'].median()
titanic_df['Age'] = titanic_df['Age'].fillna(median)
```

```python
#  fill the embarked column with mode of the data
mode = titanic_df['Embarked'].mode()
titanic_df['Embarked'].fillna(mode)
```

```python
titanic_df.isnull().sum()
```

```python
#  hence all missing values sorted
```

```python
#  lets do multivariate analysis - analyzing each varible for its contribution and role
```

```python
# plot 3 plots in each row, set the design using plotlib
fig, axes = plt.subplots(2, 3, figsize=(18,12))

# draw sns plots
# for categorical values we use count plots and for numerical values we use histplots or kernel density plots
sns.countplot(ax=axes[0, 0], x='Survived', data=titanic_df).set_title('Survival Distribution')
sns.countplot(ax=axes[0, 1], x='Pclass', data=titanic_df).set_title('Survival Distribution')
sns.countplot(ax=axes[0, 2], x='Sex', data=titanic_df).set_title('Survival Distribution')
sns.countplot(ax=axes[1, 0], x='SibSp', data=titanic_df).set_title('Survival Distribution')
sns.countplot(ax=axes[1, 1], x='Parch', data=titanic_df).set_title('Survival Distribution')
sns.countplot(ax=axes[1, 2], x='Embarked', data=titanic_df).set_title('Survival Distribution')
```

```python
# **Key Insights (Categorical):**
# - **Survival:** Most passengers (over 500) did not survive.
# - **Pclass:** The 3rd class was the most populated, followed by 1st and then 2nd.
# - **Sex:** There were significantly more males than females.
# - **Embarked:** The vast majority of passengers embarked from Southampton ('S').
# - **SibSp & Parch:** Most passengers traveled alone.
```

```python
#  plotting the histplot for age and fair
fig, axes = plt.subplots(1,2, figsize=(16,6))
sns.histplot(ax = axes[0], data = titanic_df, x = 'Age', kde=True,bins = 30).set_title('age_distribution')
sns.histplot(ax=axes[1], data=titanic_df, x='Fare', kde=True, bins=40).set_title('Fare Distribution')

plt.show()
```

```python
#  we will do bivariate analysis -> comparing one variable against other , here feature vs survival
fig, axes = plt.subplots(2,2, figsize=(16,12))

# Pclass vs. Survived
sns.barplot(ax=axes[0, 0], x='Pclass', y='Survived', data=titanic_df).set_title('Survival Rate by Pclass')

# Sex vs. Survived
sns.barplot(ax=axes[0, 1], x='Sex', y='Survived', data=titanic_df).set_title('Survival Rate by Sex')

# Embarked vs. Survived
sns.barplot(ax=axes[1, 0], x='Embarked', y='Survived', data=titanic_df).set_title('Survival Rate by Port')

# Has_Cabin vs. Survived
sns.barplot(ax=axes[1, 1], x='has_cabin', y='Survived', data=titanic_df).set_title('Survival Rate by Cabin Availability')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# **Key Insights (Bivariate):**
# - **Pclass:** A clear trend emerges: 1st class passengers had a >60% survival rate, while 3rd class passengers had less than 25%.
# - **Sex:** This is the strongest predictor. Females had a survival rate of ~75%, while males had a rate below 20%.
# - **Embarked:** Passengers embarking from Cherbourg ('C') had a higher survival rate than those from the other ports.
# - **Has_Cabin:** Passengers with a registered cabin number had a much higher survival rate. This is likely correlated with being in 1st class.
```

```python
# facet grid -> when plotting multiple subsets of data
```

```python
g = sns.FacetGrid(titanic_df, col='Survived', height = 6)
g.map(sns.histplot, 'Age', kde = True, bins=25)
plt.suptitle('Age Distribution by Survival Status', y=1.02)
plt.show()
```

```python
# checking for outliers
plt.figure(figsize=(10, 8))
sns.boxplot(y='Fare', data=titanic_df)
plt.title('Box Plot of Ticket Fare')
plt.ylabel('Fare')
plt.show()
```

```python
#  we will do some feature engineering stuff
# covert sibsp +parch to family sixe from there is_alone or not
# extract the surnames 

titanic_df['family_size'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df['is_alone'] = 0
titanic_df.loc[titanic_df['family_size'] == 1, 'is_alone'] = 1;
```

```python
titanic_df['is_alone']
```

```python
# Analyze the new family-related features against survival
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Survival Rate by FamilySize
sns.barplot(ax=axes[0], x='family_size', y='Survived', data=titanic_df).set_title('Survival Rate by Family Size')

# Survival Rate by IsAlone
sns.barplot(ax=axes[1], x='is_alone', y='Survived', data=titanic_df).set_title('Survival Rate for Those Traveling Alone')

plt.show()
```

```python
# Multivariate analysis - comparing multiple varibales against one other
```

```python
#  plot class-wise survival probability of male and female
#  this is multivariate ananlysis class vs gender vs survival
#  we will cat-plot for this
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=titanic_df, kind='bar',height=6, aspect=1.5)
plt.title('Survival Rate by Pclass and Sex')
plt.ylabel('Survival Probability')
plt.show()
```

```python
#  using voilin plot
plt.figure(figsize=(14, 8))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=titanic_df, split=True, palette={0: 'blue', 1: 'orange'})
plt.title('Age Distribution by Sex and Survival')
plt.show()
```

```python
# **Insight from Violin Plot:**
# - For males, the peak of the distribution for survivors (orange) is at a very young age (children), while the peak for non-survivors is in the 20-30 range.
# - For females, the distribution of survivors is much broader, indicating that females of most ages had a good chance of surviving.
```

```python
# Correlation Heatmap for numerical features
plt.figure(figsize=(14, 10))
numeric_cols = titanic_df.select_dtypes(include=np.number)
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
